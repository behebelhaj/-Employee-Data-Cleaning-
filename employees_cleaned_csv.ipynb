```python
df['Phone'] = df['Phone'].abs()
```

```python
print(df['Phone'].head())
```

```python
df['Department'] = df['Department_Region'].str.split('-').str[0]
```

```python
df['Region'] = df['Department_Region'].str.split('-').str[1]
```

```python
print(df[['Department_Region', 'Department', 'Region']].head())
```

```python
df = df.drop('Department_Region', axis=1)
```

```python
print(df.columns)
```

```python
df['Join_Date'] = pd.to_datetime(df['Join_Date'])
```

```python
print(df['Join_Date'].dtype)
```

```python
df['Age'] = df['Age'].fillna(df['Age'].median())
```

```python
df['Salary'] = df['Salary'].fillna(df['Salary'].median())
```

```python
df.groupby('Department')['Salary'].mean()
```

```python
df.groupby('Performance_Score')['Salary'].mean()
```
